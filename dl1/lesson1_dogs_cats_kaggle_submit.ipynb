```python
preds = np.argmax(log_preds, axis=1)
dog_probs = np.exp(log_preds[:,1])
```

```python
def rand_by_mask(mask): 
    #(a, size, replace)
    #generate random sample from a, size is output elements
    val = np.random.choice(np.where(mask)[0], 4, replace=False)
    return val
    
def rand_by_correct(is_correct): 
    correct_flag = rand_by_mask((preds == data.val_y)==is_correct)
    return correct_flag
```

```python
def plots(images, figsize=(12, 6), rows=1, titles=None):
    f = plt.figure(figsize=figsize)
    for i in range(len(images)):
        sp = f.add_subplot(rows, len(images)//rows, i+1)
        sp.axis('Off')
        if titles is not None:
            sp.set_title(titles[i], fontsize=16)
        plt.imshow(images[i])
```

```python
def load_image_id(ds, idx):
    return np.array(PIL.Image.open(PATH+ds.fnames[idx]))

def plot_val_with_title(idxs, title):
    imgs = [load_image_id(data.val_ds, x) for x in idxs]
    title_probs = [dog_probs[x] for x in idxs]
    print(title)
    return plots(imgs, rows=1, titles=title_probs, figsize=(16,8))
```

```python
plot_val_with_title(rand_by_correct(True), "Correct")
```

```python
plot_val_with_title(rand_by_correct(False), "Inorrect")
```

```python
def most_by_mask(mask, mult):
    idxs = np.where(mask)[0]
    return idxs[np.argsort(mult * probs[idxs])[:4]]
```

```python
def most_by_correct(y, is_correct): 
    mult = -1 if (y==1)==is_correct else 1
    return most_by_mask((preds == data.val_y)==is_correct & (data.val_y == y), mult)
```

```python
tfms = tfms_from_model(f_model, sz, aug_tfms=transforms_side_on, max_zoom=1.1)
```

```python
def get_augs():
    batch_size = 2
    data = ImageClassifierData.from_paths(PATH, bs=batch_size, tfms=tfms, num_workers=1)
    x, y = next(iter(data.aug_dl))
    return data.trn_ds.denorm(x)[1]
```
