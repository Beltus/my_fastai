```python

f_model = resnet34
sz = 375
bs = 64
```

```python
df_indexes = df.index.values.tolist() 
```

```python
#note we dont pass in a list of ints - we pass in a length, so the indexes need to be consecutive from 0
n = len(df_indexes)-1
val_idxs = get_cv_idxs(n)
```

```python
#append .jpg to breeds to get original file names
file_names = df.breed.tolist()
file_names = [s + '.jpg' for s in file_names]

```

```python
def get_data(sz):
    tfms = tfms_from_model(f_model, sz=sz, aug_tfms=transforms_side_on)
    model_data = ImageClassifierData.from_csv(PATH, 'train', csv_fname=PATH+'labels.csv', bs=bs, tfms=tfms,
                                          suffix='.jpg', val_idxs=val_idxs, test_name='test')
    return model_data
```

```python
model_data = get_data(sz)
x, y = next(iter(model_data.val_dl))
```

```python
sz=224
```

```python
model_data = get_data(sz)
```

```python
#we are doing classification not regression
model_data.is_reg
```

```python
#defaults for multi-labels are:
#crit=binary_cross_entropy
#metrics=accuracy_thres(0.5)
hasattr(model_data, 'is_multi')
```

```python
learn = ConvLearner.pretrained(f_model, data=model_data)
```

```python
lrf = learn.lr_find()
learn.sched.plot()
```

```python
lr = 1e-1
```

```python
#n_cycle: number of epochs before lr is reset to the initial value
#cycle_mult: how the lr resets over the cycles
learn.fit(lrs=lr,n_cycle=3,cycle_mult=2)
```

```python
lrs = np.array([lr/9,lr/3,lr])
```

```python
#Make all layers trainable and set precompute to False
learn.unfreeze()
learn.fit(lrs, 3, cycle_len=1, cycle_mult=2)
```

```python
learn.sched.plot_lr()
```

```python
learn.save(f'{sz}')
```

```python
learn.sched.plot_loss()
```

```python
sz = 356
```

```python
learn.set_data(get_data(sz))
learn.freeze()
#cycle_len: Cosine Annealing scheduler - number of cycles before lr is reset to the initial value
learn.fit(lr, n_cycle=3,cycle_len=1,cycle_mult=2)
```

```python
multi_preds, y = learn.TTA(is_test=True)
test_probs = np.mean(np.exp(multi_preds), 0)
```
